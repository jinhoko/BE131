```python

from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq


import random

my_recs = []

# generate 50 sequences
for seq_num in range(50):
    # each is 50 nucleotides long
    my_seq = "" 
    #print seq_num
    #print 45
    
    for seq_num in range(50):
        my_seq += random.choice("ATGC")
      #print seq_num
        my_recs.append(SeqRecord(seq=Seq(my_seq), id="seq%d" % seq_num)) # note each base has an attribute as per the obeject

SeqIO.write(my_recs, open("output.fasta", "w"), "fasta")

print (my_seq)
print(my_recs)

# Convert FASTQ and FASTA

#for seq_rec in SeqIO.parse(input.fastq, fastq):
#    my_recs.append(seq_rec)

#SeqIO.write(my_recs, open(output.fasta, w), fasta)

#print my_recs
```

```python
#INsert into genes
import sqlite3
conn = sqlite3.connect('my.db')
c = conn.cursor()

for i in range(12):  
    c.execute("""INSERT INTO genes (id, name, EnId) VALUES ('"""+ str(i) + """' ,"Gene",'"""+ str(11-i) + """');""")
    
conn.commit()
```

```python
#Insert into enzymes table
import sqlite3
conn = sqlite3.connect('my.db')
c = conn.cursor()

for i in range(12):  
    c.execute("""INSERT INTO enzyme (id, name, PaId) VALUES ('"""+ str(i) + """' ,"Enzyme x",'"""+ str(i%4) + """');""")
    
conn.commit()
```
